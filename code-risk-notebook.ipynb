# Identifying Risk In A Large Codebase

```bash
# Check out a FOSS git repo to use as an example:

cd /tmp/

git clone https://github.com/wikimedia/mediawiki.git
```

## `gource` Provides a 1000-foot view of developer activity in SCM

Install instructions here: [gource.io](http://gource.io/)


```bash
gource /tmp/mediawiki
```

## `cloc` Helps Answer The Question: Just How Large _Is_ This Codebase?

You can install `cloc` with any package manager (eg homebrew, yum, apt, cygwin).

```bash
cloc .
```

### What kind of language is used to describe the codebase?

```bash
bin/human /tmp/mediawiki
```

## What are the most popular files in the codebase?

```bash
bin/git_most_popular_files /tmp/mediawiki 28 days ago
```

### Drill-down: How many lines are in that file that the devs love so much?

Here I am using `wc` which is a built-in command on Mac, Linux and Cygwin.

```bash
wc -l /tmp/mediawiki/includes/MediaWikiServices.php
```

```bash
cloc /tmp/mediawiki/includes/MediaWikiServices.php
```

### Who are the most prolific authors?

```bash
bin/git_authors_leaderboard /tmp/mediawiki 28
```

### Who _Deletes_ The Most Code?

```bash
bin/git_leaderboard_by_deletions /tmp/mediawiki 28
```

At Wikipedia, a _bot_ deletes the most code. That's _fascinating!_
